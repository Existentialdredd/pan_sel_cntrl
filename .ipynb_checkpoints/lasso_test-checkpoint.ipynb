<h2>Estimator With Only ScikitLearn Lasso</h2>

```python
dp = data_pan
di = data_inst
de = data_err

dep_nm = inpt['en_nm']
ex_nm = inpt['ex_nm']
n_exo = inpt['n_exo']
ins_nm = list(data_inst.columns)[1:]
adep_nm = [ 0 for i in range(len(dep_nm))]
epsil = 0.05
ain_cf = [ 0 for i in range(len(dep_nm))]
ain_cf_rm = [ 0 for i in range(len(dep_nm))]
excf1 = []

for g in range(len(ex_nm)):
    for k in range(inpt['ncs']):
        dsl = dp.loc[dp['crs']==k+1,['t'] + [dep_nm[g]] + ex_nm]
        dsl = pd.merge(dsl,di,how ="inner", on='t')
        lin_reg = linear_model.LinearRegression()
        lin_reg.fit(dsl.loc[:,ex_nm+ins_nm].values,dsl.loc[:,dep_nm[g]].values.reshape(dsl.shape[0],1))
        excf1.append([lin_reg.coef_[0][i] for i in range(inpt['n_exo'])])

    incf = [0 for i in range(inpt['t_inst'])]    
    excf = [np.mean([excf1[i][j] for i in range(len(excf1))] ) for j in range(n_exo)]

    adep_nm[g] = ''.join(['a',dep_nm[g]])
    dp[adep_nm[g]] = (dp.loc[:,dep_nm[g]].values.reshape(dp.shape[0],1)
                     -dp.loc[:,ex_nm].values.dot(np.array(excf).reshape(inpt['n_exo'],1)))
    
    ain_cf1= []
    for k in range(inpt['ncs']):
        ds2 = dp.loc[dp['crs']==k+1,['t'] + [adep_nm[g]] ]
        ds2 = pd.merge(ds2,data_inst,how ="inner", on='t')
        lasso_reg = linear_model.Lasso(alpha = 0.4)
        lasso_reg.fit(ds2.loc[:,ins_nm].values,ds2.loc[:,adep_nm[g]].values.reshape(ds2.shape[0],1))
        ain_cf1.append(list(lasso_reg.coef_))


    ain_cf[g] = excf + [np.mean([ain_cf1[i][j] for i in range(len(ain_cf1)) 
                                             if np.abs(ain_cf1[i][j]) > epsil])
                                             for j in range(len(ain_cf1[0]))] 

    ain_cf_rm[g] = [[1,1] + [int(np.abs(ain_cf1[j][i])>epsil) 
                            for i in range(len(ain_cf1[0]))] 
                            for j in range(len(ain_cf1))]
```

<h2>Estimator With Both My Soft Thresholding and ScikitLearn Lasso</h2>

```python
dl = data_pan.loc[data_pan['crs']==1,['t','Z1,1','Z2,1','Z2,2']]
dps = data_pan.copy()
dl = pd.merge(dl,data_inst,how ="inner", on='t')

alph = 1
lamb = 0.5

dep_nm = ['Z1,1']
reg_nm = inpt['ex_nm']
ins_nm = list(data_inst.columns)[1:]
excoeff1 = []

for k in range(inpt['ncs']):
    dl = data_pan.loc[data_pan['crs']==k+1,['t'] + dep_nm + reg_nm]
    dl = pd.merge(dl,data_inst,how ="inner", on='t')
    lin_reg = linear_model.LinearRegression()
    lin_reg.fit(dl.loc[:,reg_nm+ins_nm].values,dl.loc[:,dep_nm].values.reshape(dl.shape[0],1))
    excoeff1.append([lin_reg.coef_[0][i] for i in range(inpt['n_exo'])])

incoeff = [0 for i in range(inpt['t_inst'])]    
excoeff = [np.mean([excoeff1[i][j] for i in range(len(excoeff1))] ) for j in range(2)]

adep_nm = ''.join(['a',dep_nm[0]])
dps[adep_nm] = dps.loc[:,dep_nm] - dps.loc[:,reg_nm].values.dot(np.array(excoeff).reshape(inpt['n_exo'],1))
```

```python
all_1_incoeff = []
all_2_incoeff = []
for k in range(inpt['ncs']):
    pdl = dps.loc[data_pan['crs']==k+1,['t'] + [adep_nm] + reg_nm ]
    dl = pd.merge(pdl,data_inst,how ="inner", on='t')
    for l in range(500):
        for j in range(len(incoeff)):
            pincoeff = np.array([incoeff[i] for i in range(len(incoeff)) if i!=j]).reshape(len(incoeff)-1,1)
            pres_nm = [ins_nm[i] for i in range(len(incoeff)) if i!=j]
            presid = (dl.loc[:,adep_nm].values.reshape(dl.shape[0],1)
                         - dl.loc[:,pres_nm].values.dot(pincoeff))
            presid = 1/dl.shape[0]*dl.loc[:,ins_nm[j]].T.dot(presid)
            sft_val = np.sign(presid)*(np.abs(presid) - lamb*alph)
            sft = np.sign(presid)*(np.abs(presid) - lamb*alph)*int(np.abs(presid) - lamb*alph>0)/(1+lamb*(1-alph))
            incoeff[j] = sft[0]     
    all_1_incoeff.append(list(incoeff)) 

    lasso_reg = linear_model.Lasso(alpha = 0.4,fit_intercept=False)
    lasso_reg.fit(dl.loc[:,ins_nm].values,dl.loc[:,adep_nm].values.reshape(dl.shape[0],1))
    all_2_incoeff.append(list(lasso_reg.coef_))

epsil = 0.05
all_11_incoeff = [np.mean([all_1_incoeff[i][j] for i in range(len(all_1_incoeff)) 
                                               if np.abs(all_1_incoeff[i][j]) > epsil])
                                               for j in range(len(all_1_incoeff[0]))] 
all_12_incoeff = [ [int(np.abs(all_1_incoeff[j][i])>epsil) 
                                               for i in range(len(all_1_incoeff[0]))] 
                                               for j in range(len(all_1_incoeff))]

all_21_incoeff = [np.mean([all_2_incoeff[i][j] for i in range(len(all_2_incoeff)) 
                                              if np.abs(all_2_incoeff[i][j]) > epsil])
                                              for j in range(len(all_2_incoeff[0]))] 
all_22_incoeff = [ [int(np.abs(all_2_incoeff[j][i])>epsil) 
                                               for i in range(len(all_2_incoeff[0]))] 
                                               for j in range(len(all_2_incoeff))]
```
